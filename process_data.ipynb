```python
import os
```

```python
data_dir = r'data'
celeb_real_dir = os.path.join(data_dir, 'Celeb-real')
celeb_fake_dir = os.path.join(data_dir, 'Celeb-synthesis')
```

```python
real_to_fake_dict = {}
```

```python
# Iterate through directory using os.scandir() (faster than listdir)
for entry in os.scandir(celeb_real_dir):
    real_to_fake_dict[entry.name] = []

real_to_fake_dict
```

```python
# Iterate through directory using os.scandir() (faster than listdir)
for entry in os.scandir(celeb_fake_dir):
    s = os.path.splitext(entry.name)[0].split('_')
    real_video_name = s[0] + '_' + s[2] + '.mp4'
    real_to_fake_dict[real_video_name].append(entry.name)
```

```python
real_to_fake_dict
```

```python
import os
import shutil

# Paths where real and fake videos are stored
real_videos_path = r"data\Celeb-real"
fake_videos_path = r"data\Celeb-synthesis"
output_directory = r"dataset2"  # Where the new structured folders will be created


# Ensure the output directory exists
os.makedirs(output_directory, exist_ok=True)

# Iterate through the real videos and their corresponding fake videos
for real_video, fake_videos in real_to_fake_dict.items():
    # Create a directory named after the real video (without extension)
    real_video_name = os.path.splitext(real_video)[0]  # Remove extension
    real_video_folder = os.path.join(output_directory, real_video_name)
    os.makedirs(real_video_folder, exist_ok=True)

    # Move the real video to the new folder
    real_video_src = os.path.join(real_videos_path, real_video)
    real_video_dest = os.path.join(real_video_folder, real_video)
    if os.path.exists(real_video_src):
        shutil.move(real_video_src, real_video_dest)
        print(f"Moved real video: {real_video_src} -> {real_video_dest}")
    else:
        print(f"Real video not found: {real_video_src}")

    # Move each fake video to the same folder
    for fake_video in fake_videos:
        fake_video_src = os.path.join(fake_videos_path, fake_video)
        fake_video_dest = os.path.join(real_video_folder, fake_video)
        if os.path.exists(fake_video_src):
            shutil.move(fake_video_src, fake_video_dest)
            print(f"Moved fake video: {fake_video_src} -> {fake_video_dest}")
        else:
            print(f"Fake video not found: {fake_video_src}")

print("File organization complete!")
```

```python
import os
import cv2
import numpy as np

dataset_root = "dataset"
output_root = "processed_dataset_10_Frames"  # Where extracted frames will be stored

os.makedirs(output_root, exist_ok=True)

def extract_and_save_frames(video_path, save_dir, num_frames=10):
    """Extracts frames from a video and saves them as images."""
    cap = cv2.VideoCapture(video_path)
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    frame_indices = list(map(int, np.linspace(0, total_frames - 1, num_frames)))

    os.makedirs(save_dir, exist_ok=True)

    frame_id = 0
    for idx in frame_indices:
        cap.set(cv2.CAP_PROP_POS_FRAMES, idx)  # Jump to the required frame
        ret, frame = cap.read()
        if not ret:
            break
        frame_path = os.path.join(save_dir, f"frame_{frame_id}.jpg")
        cv2.imwrite(frame_path, frame)
        frame_id += 1
    
    cap.release()

# Iterate through all videos and extract frames
for folder in os.listdir(dataset_root):
    video_folder = os.path.join(dataset_root, folder)
    if os.path.isdir(video_folder):
        for video in os.listdir(video_folder):
            if video.endswith(".mp4"):
                video_path = os.path.join(video_folder, video)
                save_path = os.path.join(output_root, folder, os.path.splitext(video)[0])  # Store in new folder
                extract_and_save_frames(video_path, save_path)

print("✅ All frames extracted and saved!")
```
